```python
unsuccessful_dev = []
successful_dev = []

for dr in drs:
    print("\nProcessing ", dr, "\n")
    dr_path = os.path.join(root, dr)
    sub_drs = os.listdir(dr_path)
    for sub_dr in sub_drs:
        sub_dr_path = os.path.join(dr_path, sub_dr)
        print("At path: ", sub_dr_path)
        htk_path = os.path.join(
            "/tungstenfs/scratch/gzenke/agarmanv/data/TIMIT/",
            "htk",
            "dev",
            sub_dr
        )
        print("Corresponding htk path: ", htk_path)
        if os.path.isdir(htk_path):
            if len(os.listdir(htk_path)) == 0:
                print("No files found here, skipping.")
                unsuccessful_dev.append(sub_dr)
            else:
                htk_fnames = os.listdir(htk_path)
                for htk_fname in htk_fnames:
                    htk_srcpath = os.path.join(htk_path, htk_fname)
                    htk_dstpath = os.path.join(sub_dr_path, htk_fname)
                    shutil.copy2(htk_srcpath, htk_dstpath)
                    successful_dev.append(sub_dr)
        else:
            print("This dir does not exist, skipping.")
            unsuccessful_dev.append(sub_dr)
            
        print("\n")
```
